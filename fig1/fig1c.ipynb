```r
library(ggplot2)
library(ggthemes)
library(scales)
```

```r
precomputed_densities = read.csv("../data/precomputed_kde_fig1c_20170508.csv")
```

```r
p = ggplot()
# p = p + geom_line(data=df_filtered, aes(x=n_steps_norm_mode, color=country), stat="density", size=.8) 
p = p + geom_line(data=precomputed_densities, aes(x, y, color=country), size=.8) # new
p = p + geom_vline(xintercept = c(.5,1.5), colour="gray80", linetype = "dashed")
p = p + geom_vline(xintercept = c(1.0), colour="gray40", linetype = "longdash")
p <- p + labs(fill = "")
p <- p + labs(color = "")
p <- p + labs(linetype = "")
p <- p + xlab("Steps / Steps Mode") + ylab("Probability Density") 
p <- p + scale_colour_brewer(palette="Set1")
p <- p + scale_linetype_manual(values=c("dotdash", "dotted", "dashed", "solid"))
p <- p + theme_few(base_size=12, base_family="Helvetica")
p <- p + theme(legend.justification=c(1,.5), legend.position=c(1,.7))
p <- p + scale_x_continuous(breaks=c(0,.5,1,1.5,2,2.5,3.0), limits=c(-.1,3))

p 

ggsave(filename="../output/fig1c.pdf", plot=p, width=4, height=3)
```
