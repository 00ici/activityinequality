```r
library(ggplot2)
library(ggthemes)
```

```r
df = read.csv("../data/country_statistics_n46_20170511.csv")
```

```r
p <- ggplot(df, aes(who_2014_obesity_mean, obesity_mean,
                   label=country))
p <- p + geom_smooth(aes(group=1), method="loess",span=.9,  se = F, size=2.0, color="gray90", alpha=.1, method.args=list(family="symmetric"))
p <- p + geom_point(alpha=.2, size=3.0)

p <- p + geom_text(check_overlap = T, size=3.5)
p <- p + ylab("Fraction obese (dataset)") + xlab("Fraction obese (WHO 2014)")
p <- p + guides(color=FALSE) + guides(size=FALSE)
p <- p + theme_few(base_size=12, base_family="Helvetica")
p <- p + scale_colour_brewer(palette = "Set1")
p <- p + xlim(.0, .45) + ylim(.0, .45) 
p <-p + geom_abline(slope=1, intercept=0, color="gray10", alpha=.1, linetype="dashed")
p

ggsave(filename="../output/edfig2b.pdf", plot=p, width=4, height=3)
```

We find a significant correlation of r=0.681 between the two estimates ($p < 10^{−6}$).
